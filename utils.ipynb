```python
import cv2
import json
import os
from pathlib import Path
import sys

#ls ./FLIR_ADAS_v2/images_thermal_train/
base = "./dataset/FLIR_ADAS_v2/images_thermal_train"
json_file = "coco.json"

json_path = os.path.join(base, json_file)
print(f'Json path: {json_path}')

if not os.path.exists(json_path):
    print(f"Json file not found: {json_file}")
    sys.exit(1)   # stop here instead of continuing to open()

```

```python
with open(json_path, 'r') as f:
    data = json.load(f)



print(f"Json structure: {data.keys()}")
samples=data['annotations']
print(f"Annotation for first sample: {samples[0]}")

category=data['categories']
print(f"Categories: {category[:10]}")

image_datas=data['images']
print(f"Image data: {image_datas[0]}")

```

```python
import random
base=f"./dataset/FLIR_ADAS_v2/images_thermal_train"

def draw_bounding_boxes(image,img_name, bbox, category_id,color=None,thickness=2):
    color=color or [random.randint(0, 255) for _ in range(3)]
    x, y, w, h = bbox
    cv2.rectangle(image, (int(x), int(y)), (int(x + w), int(y + h)), color, thickness)
    cv2.putText(image, str(category_id), (int(x), int(y) - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.9, color, thickness)
    sample_output_path=os.path.join(base,img_name)
    cv2.imwrite(sample_output_path,image)
    print(f"Sample image with bounding boxes saved:{sample_output_path}")
for image in image_datas:
    if image["id"]==0:
        print(image["file_name"]) 
    img_path=os.path.join(base,image["file_name"])
    img=cv2.imread(img_path)
    img_name=Path(image["file_name"]).name
    for sample in samples:
        if sample["image_id"]==0:
            category_id=sample['category_id']
            #print(f"Sample ID: {sample['id']}")
            bbox=sample["bbox"]
            draw_bounding_boxes(img,img_name,bbox,category_id)
```
